```python
import pandas as pd

dataframe = pd.read_excel('D:/New folder/project/mydata.xlsx')
dataframe.Value = dataframe.Value.astype(int)
dataframe.Price = dataframe.price.astype(int)

dataframe.head(15)
#importing list of 15 items to select and their nutritions
```

```python
#Recursive Method for finding total value
def Recursive(Budget,Price,value,Num):
    if Num==0 or Budget==0:
        return 0
    #if we ran out of items or Budget it return 0
    if Price[Num]>Budget:
        return Recursive (Budget,Price,value,Num-1)
    #if the price of an item is more than our budget it skips it
    total_value = max(value[Num]+Recursive(Budget-Price[Num],Price,value,Num-1),Recursive(Budget,Price,value,Num-1))
    #it figures out whether it is a good choice to have item "Num" or not by finding the maximum of values of situations
    return total_value
```

```python
value = list(dataframe['Value'])
price = list(dataframe['price'])
Price = list(dataframe['Price'])
capital = 82505
#capital is the all money we have

col = capital + 1 
# total columns in knapsack array including capital = 0
row = len (value) 
# total no of rows in knapsack array for total items

ks_arr = [[0 for x in range (col)] for y in range (row)] 
#knapsack array's full of zeros yet

Recursive(capital,Price,value,15)
#it returns the maximum value that we can have this week by spending our budget
```

```python
#DP-Table as we talked about in chapter 2
for item in range (row):
    for budget in range (col):

        if price[item] > budget:
            ks_arr[item][budget] = ks_arr[item-1][budget]
            continue
            #if price of one item is more than our budget it skips it and continue


        previous_best_value = ks_arr[item-1][budget]
        
        new_best_value = value[item] + ks_arr[item-1][budget- price[item]]
        
        ks_arr[item][budget] = max(previous_best_value, new_best_value)
        #it checks if the new item can be better or not and keeps the values in DP-Table
```

```python
knapsack_value_solution = []
for row in ks_arr:
    for col in row:
        knapsack_value_solution.append(col)
max_value = max(knapsack_value_solution)
#finding knapsack maximum of value that you can have by your budget
#print ('\nFrom knapsack-DP algorithm, maximum nutritions is: ',max_value)
```

```python
knapsack_table = pd.DataFrame(ks_arr)
#print ('\nFollowing is the Knapsack table formed using DP:\n')
knapsack_table
```

```python
choosen_items = []
j = capital
for i in range (len(value)-1,0,-1):
    if ks_arr[i][j] != ks_arr[i-1][j]:
        #if the values of i and i-1 are not the same then it means i was choosen
        choosen_items.insert(0,i)
        j -= price[i]
        #it pays its price from the budget(capital)
total_items = len(choosen_items)
choosen_items
```

```python
dataframe_final = dataframe.iloc[choosen_items]

pricelist = list (dataframe_final['Price'])
total_price = 0
for object in pricelist:
    total_price += object
    #it stacks all of choosen items' prices up
    
#print ('\nTotal money used from budget of 100000 tooman in purchase is: ',total_price)
```
